```python
import pandas as pd
import re
```

```python
df=pd.read_csv('Lahore_prices_sample_data.csv')
df
```

```python
df['Price'].str.replace('Crore', '', regex=True).str.replace('Lakh', '', regex=True)
```

```python
clean_price=[]

for prices in df['Price']:
    
    if 'Crore' in prices:
        prices=float(prices.replace('Crore', ''))*10000000
        clean_price.append(prices)
        
    elif 'Lakh' in prices:
        prices=float(prices.replace('Lakh', ''))*100000
        clean_price.append(prices)
    
    else:
        prices=float(prices)
        clean_price.append(prices)
df['Price']=clean_price
df
```

```python
clean_area_unit=[]

for area_unit in df['Area']:
    
    if 'Kanal' in area_unit:
        area_in_sq_yard=float(area_unit.replace('Kanal', ''))*605
        clean_area_unit.append(area_in_sq_yard)
        
    elif 'Marla' in area_unit:
        area_in_sq_yard=float(area_unit.replace('Marla', ''))*30.2501
        clean_area_unit.append(area_in_sq_yard)
    
    else:
        area_in_sq_yard=float(area_unit)
        clean_price.append(area_in_sq_yard)
df['Area']=clean_area_unit
df
```

```python
df.to_csv('Clean_Lahore_prices_sample_data.csv')
```
